```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt                #visulization
```

```python
df = pd.read_csv("Academic-Performance-Dataset.csv")
df
```

```python
print("Shape of the dataset:", df.shape)
```

```python
print("Data types of the columns:\n", df.dtypes)
```

```python
print("Number of missing values in each column:\n", df.isna().sum())
```

```python
# Handling missing values
# Fill missing values with mean for numeric columns and forward fill for non-numeric columns
for col in df.columns:
    if df[col].dtype == 'float64' or df[col].dtype == 'int64':
        df[col].fillna(df[col].mean(), inplace=True)
    else:
        df[col].fillna(method='ffill', inplace=True)    #forwardfill
```

```python
# Create a new column 'Total Marks' by summing up marks from different subjects
df['Total Marks'] = df[['Phy_marks', 'Che_marks', 'EM1_marks', 'PPS_marks', 'SME_marks']].sum(axis=1)

# Calculate 'Percentage' based on total marks
df['Percentage'] = df['Total Marks'] / 5
df
```

```python
Q1 = df['Che_marks'].quantile(0.25)               # Detecting Outliers
Q3 = df['Che_marks'].quantile(0.75)
IQR = Q3 - Q1                                    # Interquartile Range (IQR)     

Lower_limit = Q1 - 1.5 * IQR
Upper_limit = Q3 + 1.5 * IQR

print(f'Q1 = {Q1}, Q3 = {Q3}, IQR = {IQR}, Lower_limit = {Lower_limit}, Upper_limit = {Upper_limit}')
```

```python
df[(df['Che_marks'] < Lower_limit) | (df['Che_marks'] > Upper_limit)]
```

```python
plt.figure(figsize=(10, 6))
df.boxplot(column=['Attendence', 'Phy_marks', 'Che_marks', 'EM1_marks', 'PPS_marks', 'SME_marks'])
plt.title('Boxplot of Numeric Variables')
plt.ylabel('Marks')
plt.show()
```

```python
def BinningFunction(column, cut_points, labels = None) :
    break_points=[column.min()] + cut_points + [column.max( )]
    print('Gradding According to percentage \n>60 = F \n60-70 = B \n70-80 = A\n80-100 = O')
    return pd.cut(column, bins=break_points, labels=labels, include_lowest=True)    #leftmost bin is inclusive of its left boundary.
```

```python
cut_points=[60, 70, 80]                #cut points or thresholds that define the boundaries of each bin.
labels=['F', 'B', 'A', 'O']
df['Grade']=BinningFunction(df['Percentage'], cut_points, labels)

```

```python
print("Transformed DataFrame:")
print(df)
```
